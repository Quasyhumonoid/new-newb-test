```python
#functions can invoke themselves
def hello(name):
    print("Hello",name)
    hello(name)
    #this causes and endless loop 
    #luckily there is a limit to how many times a function can repeat..
    
count=0
def limit():
    import sys
    global count 
    count += 1
    print(count,"The recursion limit is:", sys.getrecursionlimit())

#default is 3000
limit()
sys.setrecursionlimit(100)
limit()

#set a terminating condition so you don't hit the recursion limit

def decrement(num):
    while num>0:
        print(num)
        num -= 1
decrement(10)

def decrement(num):
    if num == 0:
        return
    print(num)
    decrement(num - 1)
decrement(10)

def factorial(num):
    if num < 0:
        print("factorial doesn/'t work for negative numbers")
    if num==0:
        return 1
    result = num * factorial(num-1)
    return result

print("4! =", factorial(4))
print("15! =", factorial(15))

def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)
    
print(fibonacci(8))
```

```python
#iterator: any python that responds to the next() function
#generator: creates a function that can be iterated

def generator():
    print('one')
    yield 1
    print('two')
    yield 2
    print('three')
    yield 3
g=generator()
print (g)
print(next(g))
print(next(g))
print(next(g))
# the generator needs to be reinvoked by assigning it to a variable again

def evens(limit):
    for i in range(0, limit, 2):
        yield i
        
g = evens(7)
print(next(g))
print(next(g))
g = evens(12)
evennums=list(g)
print(evennums)

def powers():
    num=0
    while True:
        num=num+1
        yield 2**num

gen=powers()
count =0
for p in gen:
    print(p)
    
    count = count+1
    if count>10:
        break
```

```python
#closures are functions nested in other functions
#they 
def gethello():
    
    def hello():
        print("suh dude")
        
    return hello

helloack = gethello()

print(helloack)
helloack()

import random

def personal(name, message):
    annotations = ['-', '*','+','=',':','^']
    annotate = random.choice(annotations)
    def greeting():
        print(annotate * 50)
        print(message, name)
        print(annotate * 50)
    return greeting

message = personal('alec', 'suck it')
message2 = personal('mary', 'whatcha want?')
message3 = personal('doug', 'getoverit')
message()
message2()
message3()

def enroll(college):
    students=[]
    def identify(student):
        students.append(student)
        print("Student,", student,", has been enrolled in", college)
        print("current students include:", students, end="\n\n")
    return identify

yale = enroll("yale")
duke = enroll("duke")

duke("alec")
yale("cart")
duke('doggo')
```

```python
#decorators add functionality to existing code
def message():
    print("decorators are cool")
    
message()

import random

def highlight(output):
    annotations = ("-","*","+",">")
    annotate = random.choice(annotations)
    print(annotate*50)
    output()
    print(annotate*50)
    return highlight
    
highlight(message)

highlight(print('dogs and stuff'))
```
